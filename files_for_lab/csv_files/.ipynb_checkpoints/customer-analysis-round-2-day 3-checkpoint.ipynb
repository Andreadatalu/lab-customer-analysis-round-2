# Round 2:

```python
import pandas as pd
import numpy as np
```

```python
#show the dataframe
data = pd.read_csv('marketing_customer_analysis.csv')
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
#Data Shape
data.shape
```

```python
#Check the columns name
print(data.head(0))
```

```python
#Rename columns
data = data.rename(columns={'EmploymentStatus':'Employment Status'})
data.head(0)
```

```python
#Numerical_columns
numerical_cols = data._get_numeric_data()
numerical_cols
```

```python
#Categorical_columns

data.select_dtypes('object')
```

```python
#Identify the NaN numbers
nulls_df = pd.DataFrame(round(data.isna().sum()/len(data),4)*100)
nulls_df
nulls_df = nulls_df.reset_index()
nulls_df
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
#Remove the NaN from several list that we considering have lower impact

data = data.dropna(subset=['State', 'Response', 'Number of Open Complaints', 'Months Since Last Claim', 'Vehicle Class', 'Vehicle Size'])
data
```

```python
#Verift if we remove the null from State', 'Response', 'Number of Open Complaints', 'Months Since Last Claim', 'Vehicle Class', 'Vehicle Size

nulls_df = pd.DataFrame(round(data.isna().sum()/len(data),4)*100)
nulls_df
nulls_df = nulls_df.reset_index()
nulls_df
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
#Change NaN Vehicle Type for Unknowna to consider this data as a valuable for the analysis

data['Vehicle Type'] = data['Vehicle Type'].fillna("Unknown")
data
```

```python
#Extract the months from the dataset and store in a separate column and rename with the name of the month

data['Month'] = pd.DatetimeIndex(data['Effective To Date']).month
data
data['Month'] = pd.to_datetime(data['Effective To Date'], format='%m/%d/%y').dt.strftime('%B')
data
```

```python
#Filter the data to show only the information for the first quarter but isn't any changes because we just have two months

data_filtered = data[data['Month'].isin(['January', 'February', 'March'])]
data_filtered
```

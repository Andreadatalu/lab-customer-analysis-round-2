# Processing for the X y split

```python
y = data2['Total Claim Amount']
X = data2.drop(['Total Claim Amount'], axis=1)
```

```python
#Separating numerical and categorical for X-features
X_cat= X.select_dtypes(include=['object'])
X_numerical = X.select_dtypes(include=[np.number])
```

```python
X_numerical.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler().fit(X_numerical)
x_normalized = transformer.transform(X_numerical)   
print(x_normalized.shape)
x_normalized 
x_normalized2 = pd.DataFrame(x_normalized, columns=X_numerical.columns)
x_normalized2
```

```python
X_numerical.describe().T
```

# Round 6

```python
#One Hot/Label Encoding (categorical). = We need to check our categorical data
X_cat = data2.select_dtypes(include = object)
```

```python
X_cat.head(20)
```

```python
#Drop the colums that we don't need it.
X_cat = X_cat.drop(['Customer','Effective To Date','Month'], axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first').fit(X_cat)

cols = encoder.get_feature_names_out(input_features=X_cat.columns)

X_cat_encode = pd.DataFrame(encoder.transform(X_cat).toarray(),columns=cols)

X_cat_encode.head()
```

```python
X_cat_encode.shape
```

```python
#Concat DataFrames
X = pd.concat([x_normalized2, X_cat_encode ], axis=1)
display(X)
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
#Running the lineal regression
from sklearn import linear_model
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
predictions = lm.predict(X_train)
r2_score(y_train, predictions)
```

```python
predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

```python
y_test[:5]
```

```python
predictions_test[:5]
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions_test)
mse
```
